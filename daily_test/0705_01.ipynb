```python
import time
import inspect

def monitor_time(func):
    def wrapper(*args, **kwargs):
        print(f"Monitoring function '{func.__name__}'...")
        
        # 获取函数的源代码
        source_lines = inspect.getsourcelines(func)[0]
        
        # 记录开始时间
        start_time = time.time()
        
        # 执行函数
        result = func(*args, **kwargs)
        
        # 计算每行代码的执行时间
        for i, line in enumerate(source_lines):
            # 忽略装饰器自身和空白行
            if "@monitor_time" in line or not line.strip():
                continue
            
            # 计算每行代码的执行时间
            line_start_time = time.time()
            exec(line)
            line_end_time = time.time()
            line_elapsed_time = line_end_time - line_start_time
            
            print(f"Line {i + 1}: {line.strip()} - Elapsed time: {line_elapsed_time:.6f} seconds")
        
        # 计算总体执行时间
        end_time = time.time()
        elapsed_time = end_time - start_time
        print(f"Total elapsed time for function '{func.__name__}': {elapsed_time:.6f} seconds")
        
        return result
    
    return wrapper

# 示例函数：计算斐波那契数列的第 n 项
@monitor_time
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

# 主程序
if __name__ == "__main__":
    result = fibonacci(30)
    print("Fibonacci(30) =", result) 
```
